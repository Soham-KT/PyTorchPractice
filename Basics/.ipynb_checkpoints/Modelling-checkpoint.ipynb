```python
import numpy as np
import seaborn as sns
import torch.nn as nn
import pandas as pd
import torch
```

```python
cars = pd.read_csv('https://gist.githubusercontent.com/noamross/e5d3e859aa0c794be10b/raw/b999fb4425b54c63cab088c0ce2c0d6ce961a563/cars.csv')
cars.head()
```

```python
sns.scatterplot(x='wt', y='mpg', data=cars)
sns.regplot(x='wt', y='mpg', data=cars)
```

```python
cars.describe()
```

```python
x_list = cars.wt.values
x_np = np.array(x_list, dtype=np.float32).reshape(-1, 1)
y_list = cars.mpg.values.tolist()
X = torch.from_numpy(x_np)
Y = torch.tensor(y_list)
```

```python
X
```

```python
Y
```

```python
w = torch.rand(1, requires_grad=True, dtype=torch.float32)
b = torch.rand(1, requires_grad=True, dtype=torch.float32)

epochs = 1000
learning_rate = 0.001
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        y_pred = X[i] * w + b
        loss_tensor = torch.pow(y_pred - Y[i], 2)
        loss_tensor.backward()
        loss_value = loss_tensor.data[0]

        with torch.no_grad():
            w -= w.grad * learning_rate
            b -= b.grad * learning_rate
            w.grad.zero_()
            b.grad.zero_()

    print(loss_value)
```

```python
print(f'Weights: {w.item()}, Bias: {b.item()}')
```

```python
y_pred = (X * w + b).detach().numpy()
y_pred
```

```python
sns.scatterplot(x=x_list, y=y_list)
sns.lineplot(x=x_list, y=y_pred.reshape(-1))
```
